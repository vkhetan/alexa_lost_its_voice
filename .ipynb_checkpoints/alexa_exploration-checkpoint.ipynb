```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="darkgrid")
```

```python
df = pd.read_excel("alexa_super_bowl.xlsx")
```

```python
df.head(2)
```

```python
# list(df)
```

```python
plt.figure(figsize=(6,6))
ax=plt.subplot(111)
ax = sns.countplot(x="Network", data=df);
plt.xticks(rotation=90);
# ax2 = sns.countplot(x = "Show", data = df)
# ax.set_xticklabels(rotation=30)
plt.title("Number of View on Each Network");
```

```python

plt.figure(figsize=(6,5))
ax = sns.countplot(x="Network", hue="Network Type", data=df)
plt.xticks(rotation=90);
plt.title("Number of View on Network(What Network Type was it?)");
```

```python
# what people were watching each hour of day
plt.figure(figsize=(10,10))
ax = sns.countplot(hue="Network", x="Hour of Day", data=df)#, palette="Set3")
plt.title("Viewership Each Hour on Different Network");
```

```python
# what netwrok people watch at what time of the day
plt.figure(figsize=(10,10))
ax = sns.countplot(hue="Network", x="Dayparts", data=df)#, palette="Set3")
plt.title("Viewership Each DayPart on Different Network");
# How many watched alexa super bown ad in various tie of a day
```

```python
# plt.figure(figsize=(12,12));
ax = sns.factorplot(x="Network", hue="Dayparts", col="Duration",
                   data=df, kind="count",
                   size=7, aspect=1.5);
plt.xticks(rotation=90);
plt.title("Viewership Each Network conditioned on Different DayPart");
# on which network people were watching alexa superbown ad
```

```python
## What network people are watchinga at different dayparts
# plt.figure(figsize=(12,12));
print("viewership at each day part time")
ax = sns.factorplot(x="Network", hue="Duration", col="Dayparts", data=df, kind="count",
                   size=5, aspect=1, col_wrap = 6, sharex = False);

loc, labels = plt.xticks();
ax.set_xticklabels(labels, rotation=90);
# ax.set_yticklabels(labels[::-1], rotation=45) # reversed order for y

# ax.set_titles("Viewership Each Network conditioned on Different DayPart(one at a time)");
```

```python
## What network people are watchinga at different dayparts
# plt.figure(figsize=(12,12));
print("viewership at each week day")

ax = sns.factorplot(x="Network", hue="Duration", col="Day of Week", data=df, kind="count",
                   size=5, aspect=1, col_wrap = 6, sharex = False);

loc, labels = plt.xticks();
ax.set_xticklabels(labels, rotation=90);
# ax.set_yticklabels(labels[::-1], rotation=45) # reversed order for y
```

```python
groups = df.groupby(['Network',"Day of Week",'Dayparts', "Hour of Day", "Duration"])
```

```python
groups.size()[:10]
```

```python
df_new = groups.size().to_frame(name = 'count').reset_index()
```

```python
df_new['Total_Duration'] = df_new['Duration']*df_new['count']
```

```python
del df_new['Duration']
```

```python
# df_new.groupby(['Network']).agg({'Day of Week' : 'nunique', 'Hour of Day' : 'nunique'})
df_new = df_new.groupby(['Network']).agg({'Day of Week' : 'nunique', 'Dayparts': 'nunique',
                                 'Hour of Day': 'nunique', 'Total_Duration': sum})
```

```python
df_new.sort_values(by='Total_Duration', ascending=False, inplace=True)
df_new.columns = ['num_day_each_week', 'num_dayparts', 'num_hour_day', 'total_duration']
```

```python
df_new
```

```python
plt.scatter(list(df_new.index), df_new['total_duration'])
plt.xticks(rotation = 90);
plt.title("Total View Duration on Each Network");
```

```python
df = df.sort_values(by='Time')
plt.figure(figsize = (8,8));
plt.plot(df['Time'], np.cumsum(df['Duration']));
plt.xticks(rotation = 90);
plt.title("Cummulative View Duration at each time stamp");
```
